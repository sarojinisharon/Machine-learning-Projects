```python
# Initializing an ANN network

def initializing_network(nn_inputs, nn_hidden, nn_outputs):
    n_network = list()
    n_hidden_layer = [{'weights':[random.uniform(-0.5,0.5) 
                                  for i in range(nn_inputs + 1)]} 
                      for i in range(nn_hidden)]
    n_network.append(n_hidden_layer)
    n_output_layer = [{'weights':[random.uniform(-0.5,0.5)
                                  for i in range(nn_hidden + 1)]} 
                      for i in range(nn_outputs)]
    n_network.append(n_output_layer)
    i= 1
    print("\n The initialised Neural Network:\n")
    for layer in n_network:
        j=1
        for sub in layer:
            print("\n Layer[%d] Node[%d]:\n" %(i,j),sub)
            j=j+1
        i=i+1
    return n_network
```

```python
# Calculating neuron activation for an input

def neuron_activate(weights, inputs):
    neuron_activation = weights[-1]
    for i in range(len(weights)-1):
        neuron_activation += weights[i] * inputs[i]
    return neuron_activation
```

```python
# Transfering neuron activation to sigmoid function 
def neuron_transfer(activation_function):
    return 1.0 / (1.0 + exp(-activation_function))
```

```python
# Forward propagating input to a network's output
def forward_propagation(n_network, row):
    inputs = row
    for layer in n_network:
        new_inputs = []
        for neuron in layer:
            neuron_activation = neuron_activate(neuron['weights'], inputs)
            neuron['output'] = neuron_transfer(neuron_activation)
            new_inputs.append(neuron['output'])
        inputs = new_inputs
    return inputs
```

```python
# Calculating the derivative of a neuron output
def transfering_derivative(output):
    return output * (1.0 - output)
```

```python
# Backpropagating error and store in neurons
def backward_propagation_error(n_network, expected):
    for i in reversed(range(len(n_network))):
        layer = n_network[i]
        errors = list()
        
        if i != len(n_network)-1:
            for j in range(len(layer)):
                error = 0.0
                for neuron in n_network[i + 1]:
                    error += (neuron['weights'][j] * neuron['delta'])
                errors.append(error)
        else:
            for j in range(len(layer)):
                neuron = layer[j]
                errors.append(expected[j] - neuron['output'])
        
        for j in range(len(layer)):
            neuron = layer[j]
            neuron['delta'] = errors[j] * transfering_derivative(neuron['output'])
 
```

```python
# Updating the network weights with error
def updating_weights(n_network, row, learning_rate):
    for i in range(len(n_network)):
        inputs = row[:-1]
        if i != 0:
            inputs = [neuron['output'] for neuron in n_network[i - 1]]
        for neuron in n_network[i]:
            for j in range(len(inputs)):
                neuron['weights'][j] += learning_rate * neuron['delta'] * inputs[j]
            neuron['weights'][-1] += learning_rate * neuron['delta']
```

```python
# Training the network for a fixed number of epochs
def training_network(n_network, train, learning_rate, n_epoch, nn_outputs):
   
    print("\n Starting Network Training:\n")
    
    for epoch in range(n_epoch):
        sum_error = 0
        for row in train:
            outputs = forward_propagation(n_network, row)
            expected = [0 for i in range(nn_outputs)]
            expected[row[-1]] = 1
            sum_error += sum([(expected[i]-outputs[i])**2 for i in range(len(expected))])
            backward_propagation_error(n_network, expected)
            updating_weights(n_network, row, learning_rate)
        print('>epoch=%d, lrate=%.3f, error=%.3f' % (epoch, learning_rate, sum_error))
    
    print("\n Network Training done:\n")
```

```python
# Forward propagating a input to a network output
def forward_propagation(n_network, row):
    inputs = row
    for layer in n_network:
        new_inputs = []
        for neuron in layer:
            neuron_activation = neuron_activate(neuron['weights'], inputs)
            neuron['output'] = neuron_transfer(neuron_activation)
            new_inputs.append(neuron['output'])
        inputs = new_inputs
    return inputs
```

```python
# Make a prediction with a network
def predict(n_network, row):
    outputs = forward_propagation(n_network, row)
    return outputs.index(max(outputs))
```
